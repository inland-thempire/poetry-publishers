```python
import requests, lxml
from bs4 import BeautifulSoup
from time import sleep
```

```python
names = []
links = []
description = []
num = 1

for i in range(0, 18):
    r = requests.get(f"https://www.clmp.org/readers/directory-of-publishers/?_dop_publication_type=chapbook%2Cpress&_dop_genre=poetry&_paged={num}")
    print(r)
    source = r.content
    soup = BeautifulSoup(source, 'html.parser')
    publishers = soup.find_all('div', class_ = 'col-lg-9')
    for p in publishers:
        names.append(p.a.text) # publisher name
        links.append(p.a['href']) # link
        try:
            description.append(p.p.text) # description
        except:
            description.append('n/a')
    num += 1
    sleep(1)
    
```

```python
import pandas as pd

df = pd.DataFrame({
    'title': names,
    'link': links,
    'description': description
})
```

```python
# saving this first CSV file

df.to_csv('pub-links-desc.csv')
```

```python
page_titles = []

for i in links:
    try:
        r = requests.get(i)
        source = r.content
        soup = BeautifulSoup(source, 'html.parser')
        page_titles.append(soup.title.text)
    except:
        pass
    sleep(2)
```

```python
page_data = []

for i in links:
    try:
        r = requests.get(i)
        source = r.content
        soup = BeautifulSoup(source, 'html.parser')
        page_data.append(soup.find('div', class_ = 'db-fields py-4').text)
    except:
        pass
    sleep(2)
```

```python
page_data[:10]
```

```python
len(page_data)
```

```python
df = pd.DataFrame({
    'title': page_titles,'data': page_data
})
```

```python
df.to_csv('clmp_publishers.csv')
```
